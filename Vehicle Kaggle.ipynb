## Classification categorical and numerical columns

```python
numeric=['price','odometer','lat', 'long', 'county_fips', 'state_fips', 'weather']
categorical=crfull.drop(numeric,axis=1)
numerical=crfull[numeric]
numerical.head()
```

## Check missing value

```python
count_missing = crfull.isnull().sum().sort_values(ascending = False)
print(count_missing)
percentage_missing = round(crfull.isnull().sum()/len(crfull)*100,2).sort_values(ascending=False)
print(percentage_missing)
```

## Drop columns Size dan Vin bacause the percentage_missing more than 50%

```python
crfull = crfull.drop(columns = ['size','vin'])
```

## Fill missing value
### for numerical using median

```python
crfull['weather'] = crfull['weather'].fillna(crfull['weather'].median())

crfull['county_fips'] = crfull['county_fips'].fillna(crfull['county_fips'].median())

crfull['odometer'] = crfull['odometer'].fillna(crfull['odometer'].median())

crfull['state_fips'] = crfull['state_fips'].fillna(crfull['state_fips'].median())
```

### for categorical using mode

```python
crfull['type'] = crfull['type'].fillna(crfull['type'].mode().values[0])

crfull['condition'] = crfull['condition'].fillna(crfull['condition'].mode().values[0])

crfull['paint_color'] = crfull['paint_color'].fillna(crfull['paint_color'].mode().values[0])

crfull['cylinders'] = crfull['cylinders'].fillna(crfull['cylinders'].mode().values[0])

crfull['drive'] = crfull['drive'].fillna(crfull['drive'].mode().values[0])

crfull['manufacturer'] = crfull['manufacturer'].fillna(crfull['manufacturer'].mode().values[0])

crfull['make'] = crfull['make'].fillna(crfull['make'].mode().values[0])

crfull['cylinders'] = crfull['cylinders'].fillna(crfull['cylinders'].mode().values[0])

crfull['county_name'] = crfull['county_name'].fillna(crfull['county_name'].mode().values[0])

crfull['state_code'] = crfull['state_code'].fillna(crfull['state_code'].mode().values[0])

crfull['fuel'] = crfull['fuel'].fillna(crfull['fuel'].mode().values[0])

crfull['transmission'] = crfull['transmission'].fillna(crfull['transmission'].mode().values[0])

crfull['year'] = crfull['year'].fillna(crfull['year'].mode().values[0])

crfull['title_status'] = crfull['title_status'].fillna(crfull['title_status'].mode().values[0])

crfull['image_url'] = crfull['image_url'].fillna(crfull['image_url'].mode().values[0])
```

## Check  missing value

```python
count_missing = crfull.isnull().sum().sort_values(ascending = False)
print(count_missing)
```

## Distribusi penjualan mobil diatas tahun 1999
#### Dapat dilihat pada barchart dibawah ini bahwa mobil paling banyak terjual pada tahun 2007

```python
years = crfull[crfull.year>1999].sort_values("year", ascending = False)
f=plt.figure(figsize=(40,15))
f.add_subplot(2,2,2)
sns.countplot(years['year'],order=years['year'].value_counts().index)
```

## Rata-rata harga tertinggi berdasarkan manufactur
#### Dapat dilihat pada barchart dibawah ini bahwa rata-rata harga tertinggi terdapat di manufactur chev

```python
prices = crfull[["manufacturer","price"]]
prices1 = prices.groupby("manufacturer").mean().sort_values('price', ascending = False).head()
prices1

plt.figure(figsize=(7,3))

x= range(5)
plt.bar(x,prices1['price'])
plt.xticks(x,prices1.index)
plt.xlabel('manufacturer')
plt.ylabel('mean price')
plt.show()
```

## Rata-rata harga tertinggi berdasarkan negara
#### Dapat dilihat pada barchart dibawah ini bahwa rata-rata harga tertinggi terdapat di pada negara Little River

```python
weath = crfull[["county_name","price"]]
weaths = weath.groupby("county_name").mean().sort_values('price', ascending = False).head()

plt.figure(figsize=(10,5))

x= range(5)
plt.bar(x,weaths['price'])
plt.xticks(x,weaths.index)
plt.xlabel('country')
plt.ylabel('price')
plt.show()
```

## Rata-rata harga tertinggi berdasarkan tipe kendaraan

## Pie Chart

#### Pie chart ini menunjukkan bahwa bahan bakar gas lebih banyak digunakan dibandingkan bahan bakar lainnya
