```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
am_losses = pd.read_csv('automobile-losses.csv')
am_risk = pd.read_csv('automobile-risk.csv')
am_spec = pd.read_csv('automobile-spec.csv')
```

## Join 3 tables using ID as key 

```python
am_join = pd.merge(am_losses, am_risk, how = 'outer', on = 'ID')
am_join2 = pd.merge(am_join, am_spec, how = 'outer', on = 'ID')
am_join2
```

## Identify missing value. If any, how will you handle it?

```python
count_missing = am_join2.isnull().sum().sort_values(ascending = False)
print(count_missing)
```

```python
#mengisi nilai missing value pada numerical dengan median
median1 = am_join2['normalized-losses'].median()
am_join2['normalized-losses'] = am_join2['normalized-losses'].fillna(median1)

median2 = am_join2['price'].median()
am_join2['price'] =am_join2['price'].fillna(median2)

median3 = am_join2['stroke'].median()
am_join2['stroke'] = am_join2['stroke'].fillna(median3)

median4 = am_join2['bore'].median()
am_join2['bore'] = am_join2['bore'].fillna(median4)

median5 = am_join2['peak-rpm'].median()
am_join2['peak-rpm'] = am_join2['peak-rpm'].fillna(median5)

median6 = am_join2['horsepower'].median()
am_join2['horsepower'] =am_join2['horsepower'].fillna(median6)

#mengisi nilai missing value pada kategorical dengan modus
modus = am_join2['num-of-doors'].mode().values[0]
am_join2['num-of-doors'] = am_join2['num-of-doors'].fillna(modus)
```

```python
count_missing = am_join2.isnull().sum().sort_values(ascending = False)
print(count_missing)
```

## Perform visualization using at least 5 difference visualization technique (barplot, scatter plot, area, boxplot, pie chart, line chart, etc)

### 1. Histogram

```python
plt.hist(am_join2['price'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10,
         cumulative=True)
plt.show()
```

```python
kategorik=['make','fuel-type','aspiration','num-of-doors', 'body-style', 'drive-wheel', 'engine-loc', 'engine-type', 'cylinder', 'fuel-system']
numerical=am_join2.drop(kategorik,axis=1)
categorical=am_join2[kategorik]
numerical.head()
```

### 2. Bar Plot

```python
f=plt.figure(figsize=(13,12))
f.add_subplot(3,3,1)
sns.countplot(numerical['normalized-losses'])
f.add_subplot(3,3,2)
sns.countplot(numerical['symboling'])
f.add_subplot(3,3,3)
sns.countplot(numerical['wheel-base'])
f=plt.figure(figsize=(13,10))
f.add_subplot(2,2,1)
sns.countplot(categorical['body-style'], order=categorical['body-style'].value_counts().index)
f.add_subplot(2,2,2)
sns.countplot(categorical['fuel-type'],order=categorical['fuel-type'].value_counts().index)
```

### 3. Scatter Plot

```python
f=plt.figure(figsize=(13,5))
f.add_subplot(1,2,1)
sns.regplot(am_join2['curb-weight'],am_join2['length'],fit_reg=True)
f.add_subplot(1,2,2)
sns.regplot(am_join2['horsepower'],am_join2['engine-size'],fit_reg=True)
```

### 4. Correlation Plot

```python
#create correlation
corr = numerical.corr(method = 'pearson')

#convert correlation to numpy array
mask = np.array(corr)

#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(20,10)
sns.heatmap(corr, 
            mask = mask, 
            vmax = 0.9, 
            square = True, 
            annot = True)
```

### 5. Box Plot

```python
f=plt.figure(figsize=(13,13))
f.add_subplot(3,1,1)
g = sns.boxplot(y=am_join2["fuel-type"], x=am_join2['price'])
```
